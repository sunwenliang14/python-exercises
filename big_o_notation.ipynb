```python
Title your chart "Big O Notation"
Label your x axis "Elements"
Label your y axis "Operations"
Label your curves or make a legend for the curves
## Curves to graph
$y = 0x + 1$ and label the curve "O(1)"
$y = log(x)$ and label the curve "O(log n)"
$y = x$ and label the curve "O(n)"
$y = x * log(x)$ and label it "O(n log n)"
$y = x^2$ and label it "O(n^2)"
$y = 2^x$ and label it "O(2^n)"
$y = n!$ and label is "O(n!)"
$y = n^n$ and label it "O(n^n)"
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.show()
```

```python
import math
x = range(1, 100)
y = [math.log(n) for n in x]
plt.plot(x, y, label='O(log n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [n + 1 for n in x]
plt.plot(x, y, label='O(1)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [n for n in x]
plt.plot(x, y, label='O(n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [x * math.log(x) for x in x]
plt.plot(x, y, label='O(n log n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [x ** 2 for x in x]
plt.plot(x, y, label='O(n^2)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [2 ** x for x in x]
plt.plot(x, y, label='O(2^n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [n ** n for n in x]
plt.plot(x, y, label='O(n^n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.yscale('linear')
plt.legend(loc='best')
plt.show()
```

```python
x = range(1, 20)
y = [math.factorial(n) for n in x]
plt.plot(x, y, label='O(n log n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.show()
```
