```python
import matplotlib.pyplot as plt
```

```python
import math
x = range(-100,101)
y = [n ** 2 - n + 2 for n in x]
plt.title("y=$x^2 -x + 2$")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.plot(x,y)
plt.annotate("Origin", xy=(0,0), xytext=(0, 125), arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
x = range(1,101)
y = [math.sqrt(n) for n in x]

plt.title("$y = \sqrt{x}$", fontsize=20)

plt.xlabel("x", fontsize=15)

plt.ylabel("$\sqrt{x}$", rotation=0, fontsize=15)

plt.plot(y)
plt.show()
```

```python
x = range(1,101)
y = [x**3 for x in x]
plt.title("$y = x^3$", fontsize=18)

plt.xlabel("x", fontsize=14)
plt.ylabel("$x^3$", rotation=0, fontsize=14)
plt.plot(y)
plt.show()
```

```python
x = range(1,10)
y = [math.tan(x) for x in x]
plt.title("$y=tan(x)$")
plt.xlabel("x")
plt.ylabel("$tan(x)$")
plt.plot(x, y)
plt.show()
```

```python
x = range(1,50)
y = [2**x for x in x]
plt.title("$y=2^x$")
plt.xlabel("x")
plt.ylabel("$y=2^x$")
plt.plot(x,y)
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(16, 6))
plt.suptitle('Subplots of the above 4 exercises', fontsize=16)
plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 2, 1)

x = range(1, 50)
y = [math.sqrt(n) for n in x]

plt.title("$y=\sqrt{x}$")
plt.xlabel("x")
plt.ylabel("y", rotation=0)

plt.plot(y)

plt.subplot(2, 2, 2)
y = [n**3 for n in x]
plt.title("$y = x^3$")
plt.ylabel("$x^3$", rotation=0)
plt.plot(y)

plt.subplot(2, 2, 3)
x = np.linspace(-2*np.pi, 2*np.pi, 500)
y = np.tan(x)

y[:-1][np.diff(y) < 0] = np.nan

plt.grid()
plt.ylim(-10, 10)
plt.xlim(-2 * math.pi, 2 * math.pi)

plt.title("$y=tan(x)$")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.plot(x, y)

plt.subplot(2, 2, 4)
x = range(1, 15)
y = [2**n for n in x]
plt.title("$y=2^x$")
plt.xlabel("x")
plt.ylabel("$y=2^x$")
plt.plot(y)

plt.show()
```

```python

x = np.linspace(0, 15, 1000)

ex1 = [math.sqrt(n) for n in x]

ex2 = [n**3 for n in x]

ex3 = [math.tan(n) for n in x]
ex3 = np.array(ex3)
ex3[:-1][np.diff(ex3) < 0] = np.nan

ex4 = [2**n for n in x]

plt.figure(figsize=(16, 8))

plt.ylim(-25, 25)
plt.xlim(0, 12)
plt.plot(x, ex1, c='blue', label='$\sqrt{x}$')
plt.plot(x, ex2, c='orange', label='$x^3$')
plt.plot(x, ex3, c='green', label='$tan(x)$')
plt.plot(x, ex4, c='purple', label='$2^x$')

plt.legend(loc='upper right')
plt.title('4 different curves on one figure')
plt.xlabel('$x$')

plt.show()
```
